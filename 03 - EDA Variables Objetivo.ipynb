Como podemos observar mediante los box plot, parece haber una gran cantidad de datos "atipicos" pero que con los gráficos no es suficiente evidencia para deshacernos de ellos.

```python
Q1 = df_complete.iloc[:, 2:27].quantile(0.25)
Q3 = df_complete.iloc[:, 2:27].quantile(0.75)
IQR = Q3 - Q1
cantidad_atipicos = 0

# Encuentra los valores atípicos y sus posiciones
outliers = {}
for col in df_complete.iloc[:, 2:27].columns:
    outlier_mask = (df_complete.iloc[:, 2:27][col] < (Q1[col] - 1.5 * IQR[col])) | (df_complete.iloc[:, 2:27][col] > (Q3[col] + 1.5 * IQR[col]))
    outliers[col] = df_complete.iloc[:, 2:27][col][outlier_mask]

for col, values in outliers.items():
    if not values.empty:
        sorted_values = values.apply(lambda x: min(abs(x - Q1[col]), abs(x - Q3[col]))).sort_values(ascending=False) #Para ver los valores por encima del Q3
        #sorted_values = values.apply(lambda x: abs(x - Q1[col])).sort_values(ascending=True) #Para ver los valores más cercanos al Q3 y fuera del Q1

        farthest_outliers = values.loc[sorted_values.index]


        print(f"Valores atípicos más alejados en la columna '{col}':")
        total_outliers = len(farthest_outliers)
        for index, value in farthest_outliers.head(10).items():
            print(f"Posición: {index}, Valor: {value}")
        print(f"Total de valores atípicos en '{col}': {total_outliers}")
        print()
        cantidad_atipicos += total_outliers

print(f"Cantidad total de datos atípicos en todas las columnas: {cantidad_atipicos}")
```

En la celda de código anterior, identificamos los valores atípicos en nuestra base de datos. Tras analizar cada columna, concluimos que estos datos atípicos no se deben a errores de medición, cálculo o datos incorrectos. Por el contrario, son fuentes únicas de información que enriquecerán los modelos, aumentando la robustez de los mismos.

##Correlación entre predictoras

```python
plt.figure(figsize=(15,9))

c = df_complete.iloc[:, 2:27].corr()
sns.heatmap(c, cmap="Blues", annot=True, fmt=".2f")
plt.show()
```

Dado el mapa de calor analizaremos la correlación de variables entre -1 y 1.

1. Correlaciones positivas: algunas variables cuentan con una correlación moderadamente alta, por ejemplo, "Carga" parece estar positivamente correlacionada con "Punto_Isoelectrico" (~0.78), lo que indica que a medida que una de estas variables aumenta, la otra tiende a hacerlo también.

2. Correlaciones cercanas a 0: podemos observar que las variables son mayormente independientes entre sí, excepto por algunas relaciones moderadas observadas (~0.50 <).

3. Correlaciones negativas: contamos con algunas variables con correlación negativa destacando, por ejemplo, "Hidrofobicidad" con "Indice_Boman" (~0.88), dandonos a entender que cuando una aumenta, la otra tiende a disminuir.

Por lo general las correlaciones entre **aminoácidos** (A, R, N, etc.) suelen ser bajas lo que nos indica que la presencia de una aminoacido específico no afecta significativamente la presencia de otro en este conjunto de datos. Sin embargo, algunas combinaciones de aminoácidos tienen correlaciones negativas o positivas moderadas, lo que podría reflejar patrones específicos de interés en la secuencia de aminoácidos.

Por otro lado las propiedades físicoquímicas como Hidrofobicidad, Indice_Boman, y Carga muestran correlaciones diversas con los diferentes aminoácidos. Por ejemplo, la correlación negativa entre Hidrofobicidad y el aminoácido R sugiere que cuando R es alto, la Hidrofobicidad tiende a ser baja.
