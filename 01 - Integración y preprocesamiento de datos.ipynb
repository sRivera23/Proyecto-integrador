```python
def fasta_to_dataframe(fasta_file):
    records = SeqIO.parse(fasta_file, "fasta")
    data = []
    for record in records:
        data.append({"id": record.id, "sequence": str(record.seq)})
    return pd.DataFrame(data)

star_pep = f'{starpep_path}/Peptides.fasta'
no_amp = f'{unal_path}/data_neg.fasta'
unal = f'{unal_path}/multilabel_CSV.csv'
star_car = f'{starpep_path}/Function_metadata.csv'


star_pep_df = fasta_to_dataframe(star_pep) #DF de ID y secuencia de péptidos Starpep
no_amp_df = fasta_to_dataframe(no_amp) #DF No Antimicrobianos UNAL
unal_df = pd.read_csv(unal) #DF UNAL completo
starpep_car = pd.read_csv(star_car) #DF Starpep carácteristicas
```

##Ajuste de péptidos antimicrobianos

```python
def actualizar_antimicrobiano(df):
    # Verificar si alguna de las columnas antimicrobianas tiene un valor de 1
    condiciones_antimicrobianas = (
        (df['Antifúngico'] == 1) |
        (df['Antiviral'] == 1) |
        (df['Anti Gram +'] == 1) |
        (df['Anti Gram -'] == 1) |
        (df['Antibacteriano'] == 1) |
        (df['AntiVIH'] == 1) |
        (df['Antiparasitario'] == 1)
    )

    # Asignar 1 a la columna 'Antimicrobiano' donde se cumplan las condiciones
    df.loc[condiciones_antimicrobianas, 'Antimicrobiano'] = 1

    return df
```

```python
df_final = actualizar_antimicrobiano(df_final)
```

##Filtrado de péptidos con aminoácidos no naturales y péptidos con variación menor a 3 en su composición.

```python
# Función para detectar aminoácidos no válidos
def contains_invalid_amino_acids(sequence):
    valid_amino_acids = set('ACDEFGHIKLMNPQRSTVWY')  # Conjunto de aminoácidos estándar
    return any(aa not in valid_amino_acids for aa in sequence)

# Función para contar aminoácidos únicos en una secuencia
def unique_amino_acids(seq):
    return len(set(seq)) >= 3

def process_dataframe(df):
    # Filtrar secuencias con aminoácidos válidos
    df['Valid'] = df['sequence'].apply(lambda seq: not contains_invalid_amino_acids(seq))

    # Filtrar secuencias con al menos 3 aminoácidos diferentes
    df['diversity'] = df['sequence'].apply(unique_amino_acids)

    # Conservar las secuencias válidas y con suficiente diversidad
    valid_df = df[(df['Valid']) & (df['diversity'])].drop(columns=['Valid', 'diversity'])

    # Secuencias no válidas o con poca diversidad
    invalid_df = df[~((df['Valid']) & (df['diversity']))]

    return valid_df, invalid_df

valid_df1, invalid_df1 = process_dataframe(df_final)
valid_df2, invalid_df2 = process_dataframe(df_final_positive)
```

##Verificación y eliminación de péptidos con una secuencia menor a 7 y mayor a 100

```python
# Definir la función para evaluar la longitud de las secuencias
def evaluar_secuencia(secuencia):
    return len(secuencia)

# Función para analizar las secuencias de un DataFrame
def analizar_secuencias(df):
    menoresSiete = 0
    mayoresCien = 0
    posiciones = []

    for index, secuencia in df['sequence'].items():
        valor = evaluar_secuencia(secuencia)
        if valor > 100:
            mayoresCien += 1
            posiciones.append(index)
        elif valor < 7:
            menoresSiete += 1
            posiciones.append(index)

    return menoresSiete, mayoresCien, posiciones

# Aplicar la función a valid_df1
menoresSiete_df1, mayoresCien_df1, posiciones_df1 = analizar_secuencias(valid_df1)
print(f"Resultados para valid_df1:")
print(f"Cantidad de secuencias menores a 7: {menoresSiete_df1}")
print(f"Cantidad de secuencias mayores a 100: {mayoresCien_df1}")
print(f"Posiciones de cada péptido con secuencias menores a 7 o mayores a 100 en valid_df1: {posiciones_df1}")

# Aplicar la función a valid_df2
menoresSiete_df2, mayoresCien_df2, posiciones_df2 = analizar_secuencias(valid_df2)
print(f"\nResultados para valid_df2:")
print(f"Cantidad de secuencias menores a 7: {menoresSiete_df2}")
print(f"Cantidad de secuencias mayores a 100: {mayoresCien_df2}")
print(f"Posiciones de cada péptido con secuencias menores a 7 o mayores a 100 en valid_df2: {posiciones_df2}")

# Crear DataFrames filtrados para ambos
df_secuencias_filtradas_df1 = valid_df1.loc[posiciones_df1]
df_secuencias_filtradas_df2 = valid_df2.loc[posiciones_df2]
```

```python
secuencias_filtradas = df_secuencias_filtradas_df1['sequence']
secuencias_filtradas2 = df_secuencias_filtradas_df2['sequence']

valid_df1_cleaned = valid_df1[~valid_df1['sequence'].isin(secuencias_filtradas)]
valid_df2_cleaned = valid_df2[~valid_df2['sequence'].isin(secuencias_filtradas2)]
```
